```python
# def merupakan keyword dalam python yang berfungsi untuk mendefinisikan fungsi
# fuctinon name
# function parameter berguna untuk memasukkan data dalam fungsi, ini bersifat opsional
def multiple (x, y):
  print("Multiply {} and {}". format (x,y))
  result = x * y
# function code merupakan barisan kode
  return result
# function return statemen berfungsi memberikan nilai tertentu
```

```python
def penjumlahan(a, b):
  c= a +b
  return c
```

```python
# memanggil fungsi penjumlahan
# dengan memaukkan nilai a = 10 dan b = 5
penjumlahan(10, 5)
```

```python
# memanggil fungsi penjumlahan
# dengan memasukkan nilai a = 10 dan b = 5

hasil_penjumlahan = penjumlahan(10, 5)
hasil_penjumlahan
```

```python
# mendefinisikan fungsi penjumlahan
# dengan parameter a dan b
# outputnya adalah hasil jumlah dari a dan b
# dengan b = 10 sebagai nilai awal

def penjumlahan(a,b=10):
  c= a +b
  return c
```

```python
# memanggil fungsi penjumlahan
# dengan memasukkan nilai a = 10

hasil = penjumlahan(10)
hasil
```

```python
# memanggil fungsi penjumlahan
# dengan memasukkan nilai a = 10 dan b = 5

hasil = penjumlahan(10, 5)
hasil
```
