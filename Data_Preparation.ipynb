```python
#  masukkan link pada variabel link data
link_data= "https://bit.ly/spending_vs_profile"
# baca file dari link data
raw_data= pd.read_csv(link_data)
# lihat data
raw_data
```

Membuang kolom dari dataframe

```python
raw_data = raw_data.drop(['brth_cntry'], axis = 1)

# lihat data
raw_data
```

Menguabah index dari data frame

```python
# menggunakan kolom hh_id sebagai index dari tabel
raw_data = raw_data.set_index('hh_id')
raw_data
```

```python
# Jika ingin mengembalikan kembali bisa pakai
raw_data = raw_data.reset_index()
raw_data
```

Melakukan filtering dari data

```python
# Ambil data dari raw data yang nilai spendingnya diatas 400
raw_data[raw_data['spending'] > 400]
```

```python
# bisa juga dengan method query untuk melakukan filtering
raw_data.query("spending > 400")
```

```python
# buat dummy variable dari variabel ktegori
pd.get_dummies(raw_data['occupation'])
```

```python
pd.get_dummies(raw_data[['children', 'employment']])
```

```python
# buat dummy variables dari variable kategorik dengan membuang kategori pertama dari ata
pd.get_dummies(raw_data[['children', 'employment']], drop_first=True)
```

```python
# lakukan pivoting untuk mengubah bentuk ke bentuk lainnya. yaitu dari long ke wide/wide ke long
raw_data[['hh_id', 'income', 'spending']].pivot_table(
    index='hh_id',
    columns='income',
    values = 'spending'
)
```
